```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager as fm
import numpy as np
import random as rd

vi = np.loadtxt("virus_tr.txt")
vi2 = np.loadtxt("vi2.txt")
vi3 = np.loadtxt("vi3.txt")
X = np.r_[vi2,vi3]
X = np.r_[X,vi]

m=X.shape[0] 
n=X.shape[1] 
K=3
```

```python
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_samples, silhouette_score

cnt=0;
cnt_t=0
iter=100
for i in range(iter):
    gmm = GaussianMixture(n_components=3, random_state=i)
    gmm_labels = gmm.fit_predict(X)
    
    print(silhouette_score(X,gmm_labels))
    
    if silhouette_score(X,gmm_labels)>cnt:
        cnt=silhouette_score(X,gmm_labels)
        cnt_t=i
    plt.scatter(X[:, 0], X[:, 1],c=gmm_labels, s=40, cmap='viridis');
print("avr",cnt,"i",cnt_t)
```

```python
from Bio import SeqIO
records = list(SeqIO.parse("tt.fastq", "fastq"))
Y={}
C=gmm_labels+1 
for k in range(K):
        Y[k+1]=np.array([]).reshape(136,0)
for i in range(m):
        Y[C[i]]=np.c_[Y[C[i]],X[i]]
for k in range(K):
        Y[k+1]=Y[k+1].T
Output=Y

for k in range(1,K+1): 
    print("Clustering",k,"개수 ",len(Output[k]))
    for i in range(len(Output[k])): 
        
        for a in range(100):

            if np.array_equal(vi[a],Output[k][i]) == True:

                print("Output[",k ,"][", i,"] == ",records[a].description.split(";")[3])
                              
```
