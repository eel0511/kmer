```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager as fm
import numpy as np
import random as rd
!simlord --no-sam -rr tt.fa -n 300 -mr 1500 tt
!./vectorize ./tt.fastq 4 8 100 > ./tt.txt
vi = np.loadtxt("tt.txt")

print(vi.shape)
X = vi
m=X.shape[0] 
n=X.shape[1] 
K=25 
n_iter=100
Centroids=np.array([]).reshape(n,0) 
for i in range(K): 
    rand=rd.randint(0,m-1) 
    Centroids=np.c_[Centroids,X[rand]] 
Centroids 
```

```python
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_samples, silhouette_score

# n_components로 미리 군집 개수 설정
gmm = GaussianMixture(n_components=25, random_state=1)
gmm_labels = gmm.fit_predict(X)

print(silhouette_score(vi,gmm_labels))
plt.scatter(X[:, 0], X[:, 1],c=gmm_labels, s=40, cmap='viridis');
```

```python
from Bio import SeqIO
records = list(SeqIO.parse("tt.fastq", "fastq"))
Y={}
C=gmm_labels+1 
for k in range(K):
        Y[k+1]=np.array([]).reshape(136,0)
for i in range(m):
        Y[C[i]]=np.c_[Y[C[i]],X[i]]
for k in range(K):
        Y[k+1]=Y[k+1].T
Output=Y

for k in range(1,K+1): 
    print("Clustering",k,"개수 ",len(Output[k]))
    for i in range(len(Output[k])): 
        
        for a in range(100):

            if np.array_equal(vi[a],Output[k][i]) == True:

                print("Output[",k ,"][", i,"] == ",records[a].description.split(";")[3])
                              
```

```python
from sklearn.manifold import TSNE
data = np.append(Output[1],Output[2],axis=0)
data2 = np.append(data,Output[3],axis=0)

labels  = ["cluster1" for x in range(len(Output[1]))]
labels  += ["cluster2" for x in range(len(Output[2]))]
labels  += ["cluster3" for x in range(len(Output[3]))]

labels = np.array(labels)


tsne = TSNE(learning_rate=10)
data_2d = tsne.fit_transform(data3)
print(data_2d.shape)
fig = plt.figure(figsize=(10, 10))
sns.scatterplot(data_2d[:,0], data_2d[:,1], hue=labels)
```
